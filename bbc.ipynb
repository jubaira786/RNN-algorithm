```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('./bbc_news_mixed.csv')
```

```python
data.head()
```

```python
data['label'].value_counts()
```

```python
data.shape
```

```python
data['text'].iloc[0]
```

```python
x=data['text']
```

```python
y=data[['label']]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec=TfidfVectorizer(stop_words='english')
```

```python
newdata=vec.fit_transform(x)
```

```python
newdata.toarray()
```

```python
vec.get_feature_names_out().shape
```

```python
newdata.toarray().shape
```

```python
vec.get_feature_names_out()
```

```python
df=pd.DataFrame(newdata.toarray(),columns=vec.get_feature_names_out())
```

```python
df.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
label_data=encoder.fit_transform(data[['label']])
```

```python
y=pd.DataFrame(label_data.toarray(),columns=encoder.get_feature_names_out())
```

```python
y.head()
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
model=Sequential()
model.add(Dense(1000, activation='relu',  input_shape=(df.shape[1],)))
model.add(Dense(1000, activation='relu'))
model.add(Dense(5, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(df, y, test_size=0.2)
```

```python
model.fit(xtrain, ytrain, epochs=5)
```

```python
output=model.predict(xtest)
```

```python
output
```

```python
out_max=np.zeros_like(output)
for i in range(output.shape[0]):
    out=np.argmax(output[i])
    out_max[i][out]=1
```

```python
out_max=encoder.inverse_transform(out_max)
```

```python
out_max
```

```python
ytest=encoder.inverse_transform(ytest)
```

```python
ytest
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest, out_max))
```

```python
model.save('bbc_model.h5')
```
